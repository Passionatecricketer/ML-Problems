```python
def cat(cat): 
    if cat == 'Fair':
        return 1
    elif cat == 'Good':
        return 2
    elif cat == 'Very Good':
        return 3
    elif cat == 'Premium':
        return 4
    elif cat == 'Ideal':
        return 5
    else:
        return 6
df['cut'] = df['cut'].apply(cat)
```

```python
def cat(cat): 
    if cat == 'I1':
        return 1
    elif cat == 'SI1':
        return 2
    elif cat == 'SI2':
        return 3
    elif cat == 'VS1':
        return 4
    elif cat == 'VS2':
        return 5
    elif cat == 'VVS1':
        return 6
    elif cat == 'VVS2':
        return 7
    elif cat == 'IF':
        return 8
    else:
        return 9
df['clarity'] = df['clarity'].apply(cat)
```

```python
X = df.drop("price",axis=1)
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
